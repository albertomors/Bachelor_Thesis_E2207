```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def save_plot(file_name):
    path = f'figures/{file_name}.png'
    plt.savefig(path, bbox_inches='tight', dpi=400, transparent = False, facecolor = 'w')

def save_plot_svg(file_name):
    path = f'figures/{file_name}.svg'
    plt.savefig(path, format = 'svg', bbox_inches='tight', dpi=400, transparent = False, facecolor = 'w')
```

```python
# 19 microsecond sample time

sampleTime = 62e-6
t_start = 60000;
t_stop = t_start + 250#int(1/sampleTime)
t_span = (t_stop - t_start)*sampleTime


with open('stable.txt', 'r') as f:
    lines = f.read().splitlines()

    lines = lines[t_start:t_stop]

    values = np.zeros((len(lines),6))
    #values.reshape(len(lines),6)
    for i in range(len(lines)):
        tempvalues = lines[i].split(' ')
        for j in range(len(tempvalues)):
            values[i,j] = round(float(tempvalues[j]),0)


t = np.linspace(t_start,t_stop,len(values))       
```

```python
uc = np.array([values[:,3],values[:,4],values[:,5]])

um = [list(), list(), list(), list()]

u1 = 0
u2 = 0
u3 = 0
u4 = 0

for i in range(len(uc[0])):
    
    ux = uc[0][i]
    uy = uc[1][i]
    uz = abs(uc[2][i])


    if ux >= 0:
        u1 = (ux - uz) 
        u3 = (ux + uz)

        if u1 < 0:
            u1 = 255 + u1

        if u3 < 0:
            u3 = 255 + u3

        u3 = -u3

    elif ux < 0:
        ux = abs(ux)
        u1 = (ux + uz)
        u3 = (ux - uz)

        if u1 < 0:
            u1 = 255 + u1

        if u3 < 0:
            u3 = 255 + u3

        u1 = -u1

    if uy >= 0:
        u2 = (uy - uz)
        u4 = (uy + uz)

        if u2 < 0:
            u2 = 255 + u2

        if u4 < 0:
            u4 = 255 + u4

        u4 = -u4

    elif uy < 0:
        uy = abs(uy)
        u2 = (uy + uz)
        u4 = (uy - uz)


        if u2 < 0:
            u2 = 255 + u2

        if u4 < 0:
            u4 = 255 + u4

        u2 = -u2
        
    um[0].append(u1)
    um[1].append(u2)
    um[2].append(u3)
    um[3].append(u4)
    

```

```python
fig,axs = plt.subplots(2,2, dpi=800, figsize = (20,8))

axs[0,0].plot(t,um[0])
axs[0,0].set_title('Solenoid X_A')
axs[0,0].grid()

axs[0,1].plot(t,um[1])
axs[0,1].set_title('Solenoid Y_A')
axs[0,1].grid()

axs[1,0].plot(t,um[2])
axs[1,0].set_title('Solenoid X_B')
axs[1,0].grid()

axs[1,1].plot(t,um[3])
axs[1,1].set_title('Solenoid Y_B')
axs[1,1].grid()

save_plot_svg('bangbang')
plt.show()
```
